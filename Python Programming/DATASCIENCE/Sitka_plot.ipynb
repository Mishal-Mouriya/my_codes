```python
import matplotlib.pyplot as plt
from pathlib import Path
import csv
from datetime import datetime as dt

# Read CSV file
csv_file = Path(r"D:\heavens feel\PYTHON RESOURCES\3rd edition resources\pcc_3e-main\chapter_16\the_csv_file_format\weather_data\sitka_weather_2021_simple.csv")
lines = csv_file.read_text().splitlines()
info = csv.reader(lines)

# Show header row
for index, header in enumerate(next(info)):
    print(index, header)

# Parse data
high = []
low = []
current_date = []

for row in info:
    try:
        high.append(int(row[4]))
        low.append(int(row[5]))
        current_date.append(dt.strptime(row[2], "%Y-%m-%d"))
    except ValueError:
        # Skip rows with missing or bad data
        print(f"Missing data for row: {row}")
        continue

# Plotting
fig, ax = plt.subplots()

ax.set_title("Weather Highs and Lows", fontsize=18)
ax.set_xlabel("Dates", fontsize=14)
ax.set_ylabel("Temperatures (F)", fontsize=14)

ax.plot(current_date, high, color='red', label="High")
ax.plot(current_date, low, color='blue', label="Low")
ax.legend()

fig.autofmt_xdate()  # rotates the x-axis dates
plt.show()
```

```python
import matplotlib.pyplot as plt
from pathlib import Path
import csv
from datetime import datetime as dt

# Read CSV file
csv_file = Path(r"D:\heavens feel\PYTHON RESOURCES\3rd edition resources\pcc_3e-main\chapter_16\the_csv_file_format\weather_data\sitka_weather_2021_simple.csv")
lines = csv_file.read_text().splitlines()
info = csv.reader(lines)

# Show header row
for index, header in enumerate(next(info)):
    print(index, header)

# Parse data
high = []
low = []
current_date = []

for row in info:
    try:
        high.append(int(row[4]))
        low.append(int(row[5]))
        current_date.append(dt.strptime(row[2], "%Y-%m-%d"))
    except ValueError:
        # Skip rows with missing or bad data
        print(f"Missing data for row: {row}")
        continue

# Plotting
fig, ax = plt.subplots()

ax.set_title("Weather Highs and Lows", fontsize=18)
ax.set_xlabel("Dates", fontsize=14)
ax.set_ylabel("Temperatures (F)", fontsize=14)

ax.plot(current_date, high, color='red', label="High")
ax.plot(current_date, low, color='blue', label="Low")
ax.legend()

ax.fill_between(current_date ,high,low,facecolor ='pink',alpha=0.5)              #extra code to dill betwwen graph

fig.autofmt_xdate()  # rotates the x-axis dates
plt.show()
```
